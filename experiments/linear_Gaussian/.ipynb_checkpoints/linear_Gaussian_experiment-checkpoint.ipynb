```python
rng = np.random.default_rng(532124)
d = 3
n = 20
```

```python
test = LinGaussTest(rng=rng, d=d, n=n)
```

```python
samp = test.ep_post.sample(10000)
baseline = test.post.sample(50000)
```

```python
nominal_lower = np.quantile(samp, axis=0, q=[0.1, 0.2])
nominal_upper = np.quantile(samp, axis=0, q=[0.9, 0.8])


# For each l (m), and each j (d), sum over n (axis=1)
# out = covered.sum(axis=1) / n_baseline  # shape (m, d)
```

```python
covered = (baseline[None,:,:] >= nominal_lower[:,None,:]) and \
          (baseline[None,:,:] <= nominal_upper[:,None,:]) # (m, n, d)
```
